```python
lat1 = 37.
lat2 = 41.
lon1 = -102.
lon2 = -109.

longitudes = []
latitudes = []

for i in np.arange(lon2,lon1,0.0427):
    for j in np.arange(lat1,lat2,0.0427):
        longitudes.append(i)
        latitudes.append(j)
        
locations = pd.DataFrame({
    'longitude': longitudes,
    'latitude': latitudes
})
```
