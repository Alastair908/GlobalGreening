```python
import pandas as pd
import numpy as np
```

# Compiling the NDVI csvs

```python
years = [2002,2006,2010,2014,2018,2022]
```

```python
locations = pd.read_csv('../csv_folder/total_centrepoints.csv')
```

```python
for year in years:
    ndvis_2002 = pd.read_csv(f'../ndvi_csvs/ndvis2002.csv')
    ndvis_values = pd.read_csv(f'../ndvi_csvs/ndvis{year}.csv')
    ndvis_values = ndvis_values[f'{year}'] - ndvis_2002['2002']
    locations[f'{year}'] = ndvis_values
locations = locations[['longitude','latitude','2002','2006','2010','2014','2018','2022']]
locations.to_csv('../csv_folder/ndvi_results.csv')
```
